# Figure 5 - Energy Plots

```python
from wind_ae.wrapper.relax_wrapper import wind_simulation as wind_sim
from wind_ae.wrapper.wrapper_utils.plots import energy_plot

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
sim = wind_sim()
sim.load_uservars('data/HD209_13.6-2000eV_H.csv')
sim_m = wind_sim()
sim_m.load_uservars('data/HD209_mono-20eV_H.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True,figsize=[8,7])
fig.subplots_adjust(hspace=0)
energy_plot(sim.windsoln,ax=ax1,legend=False)
ax1.set_ylim(4e-11,3e-7)
# ax1.text(3.2,1e-7,'XUV',fontweight='bold')
ax1.text(1.6,1.2e-7,r'$\tau$(20eV)=1',fontsize=13)
ax1.axvline(sim.windsoln.calc_tau1_radius(20)[0],c='k')
energy_plot(sim_m.windsoln,ax=ax2)
ax2.set_ylim(4e-11,3e-7)
ax2.axvline(sim_m.windsoln.calc_tau1_radius(20)[0],c='k')
ax2.legend(loc='center right',fontsize=12)
ax2.text(1.18,1.2e-7,r'$\tau$(20eV)=1',fontsize=13)
ax2.set_xticks([1,1.16,1.57,2,3])

plt.show()
```
