```python
import subprocess

def ergs_to_wlcm(energy):
    wl_nm = 1240 / (6.242e+11*energy)
    wl_cm = wl_nm*1e-7
    return wl_cm

def ergs_to_wlnm(energy):
    wl_nm = 1240 / (6.242e+11*energy)
    return wl_nm 

def messin_around(norm_span,wndw_rslv_span,species=["H I","He I"]):
#     species = ["H I","He I"]
    spec = spectrum()
    for sps in species:
        spec.add_species(sps)
    #------Change span here-----
    #GOAL: span = spec.norm_span
#     span = np.array([1.23984254740273393e-06,9.11763351310513339e-06])*1e7 #this is span of working version, doesn't work here

#     span2 = np.array([30,91.17633513])

    #----------------------------------------------------------
    spec.set_resolved(*wndw_rslv_span) #
    spec.set_normalized(*norm_span) 
    spec.set_window(*wndw_rslv_span,kind='full')

    spec.generate(kind='full', savefile='inputs/spectrum.inp')

    print('**************************************')
    # spec.generate(kind='full', savefile='inputs/full_spectrum.inp')


    f = open("inputs/spectrum.inp", "r")
    spectral_data = f.readlines()
    spec_lines = spectral_data[8:] #first line is header
    spec_header = spectral_data[0:8]
    guess_spec_string = '#spec_prms: '
    for line in spec_header:
        guess_spec_string+=(line[:-1].split(':'))[1][1:]+','
    guess_spec_string = guess_spec_string[:-1]+'\n'
    f.close()
    f2 = open("saves/windsoln.csv","r") #risky, 
    #but can change to reference_guesses when stuff starts breaking
    guess = f2.readlines()
    guess_top = guess[0:8]
    guess_bottom = guess[80:]
    f2.close()

    f3 = open("inputs/guess.inp","w")
    f3.writelines(guess_top)
    f3.writelines(guess_spec_string)
    f3.writelines(spec_lines[0])
    for i in range(1,len(spec_lines)):
        string = '## '+spec_lines[i]
        f3.writelines(string)
    f3.writelines(guess_bottom)
    f3.close()

    new_spec = np.genfromtxt('inputs/spectrum.inp',skip_header=8,delimiter=',')
    new_wl = ergs_to_wlnm(new_spec[:,0])
    plt.scatter(new_wl,new_spec[:,1]) 
    plt.plot(new_wl,new_spec[:,1],label='new')
    # plt.show()

    #for plotting purposes
    old_spec = np.genfromtxt('inputs/reference_spec.inp',skip_header=8,delimiter=',')
    plt.scatter(ergs_to_wlnm(old_spec[:,0]),old_spec[:,1])
    plt.plot(ergs_to_wlnm(old_spec[:,0]),old_spec[:,1],label='old')
    plt.legend()
    plt.xlabel('Wavelength (nm)')
    plt.ylabel(r'w$\Phi$/F$_{UV}$') #see inputs/spectrum.inp for true header
    plt.show()

    proc = subprocess.run(["./bin/relaxed_ae"], stdout=subprocess.PIPE, text=True)
    print((proc.stdout)[-20:-2])
    # print(proc.stdout)
    if (proc.stdout)[-20:-2] == 'SAVE SOLUTION done':
        
        print("SUCESS")
        print("Number of Iterations:"+(proc.stdout)[-195:-190])
        print("SUCESS")
        print("----------------------------------------------")
        sucess = True
    else:
        print(proc.stdout)
        sucess = False
        print('norm_span',norm_span)
        print('window_span',wndw_rslv_span)
        
    return sucess
```

```python
dat = np.genfromtxt('coeffs.dat',delimiter='')
x = dat[0:18]
heat = dat[18:36]
ionizations = dat[36:54]
excitations = dat[54:]
```

```python
def Frac_deposited_as_heat(x):
    C,a,b = 0.9971, 0.2663, 1.3163
    y = C*(1 - (1 - x**a)**b )
    return y
```

```python
print(Frac_deposited_as_heat(0.3)*0.1)
print(Frac_deposited_as_heat(0.3)*0.9+)

print(Frac_deposited_as_heat(0.1))
print(Frac_deposited_as_heat(0.2))
```

```python
plt.plot(x,heat,label='heat')
plt.plot(x,Frac_deposited_as_heat(x),ls='--',label='fit')
plt.plot(x,ionizations,label='ionization')
plt.plot(x,excitations,label='excitation')
plt.xlabel('background ionization fraction (x)')
plt.ylabel('fraction of E deposited as')
plt.title(r'E$_0$ = 3 KeV')
plt.xscale('log')
plt.legend()
plt.show()
```
