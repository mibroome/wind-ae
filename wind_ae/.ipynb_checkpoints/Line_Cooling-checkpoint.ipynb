```python
def tau_finder(E_input,sim_object):
    E_ev = sim_object.windsoln.E_wl*6.24e11
    delta_E = E_ev-E_input
    r_norm = sim_object.windsoln.soln_norm['r']
    try:
        E_idx = np.where(np.abs(delta_E)==np.min(np.abs(delta_E)))[0][0]
        nspecies = len(sim_object.windsoln.sigma_wl[0])
#         taus = np.genfromtxt('taus.txt')[:,E_idx]
        taus = np.zeros_like(sim_object.windsoln.soln['r'])
        for i,col_name in enumerate(sim_object.windsoln.soln.columns[4+nspecies:4+2*nspecies]):
#             taus = simwindsoln.soln['Ncol_HI']*simwindsoln.sigma_wl[E_idx,0]+ simwindsoln.soln['Ncol_HeI']*simwindsoln.sigma_wl[E_idx,1]
            taus += sim_object.windsoln.soln[col_name]*sim_object.windsoln.sigma_wl[E_idx,i]
        tau1 = np.flip(np.array(r_norm[taus>=1]))[0]
        return tau1,E_ev[E_idx]
    except IndexError:
        return np.nan,E_ev[np.where(delta_E==np.min(np.abs(delta_E)))[0]]
```

```python
def FourPlotSpecial(planet_path='saves/windsoln.csv',line_color='k',line_style='-',alone=True,ax=0):
    
    radius_norm = 1.0
    alpha=0.5
    radius = r/Rp
    R_H = a*(Mp/(3*Mstar))**(1/3) / Rp
    
    custom_cycler = (cycler(linestyle=['-', '--', ':', '-.',(0, (1, 10)),(5, (10, 3))]))

    stack=2
    if alone==True:
        fig, ax = plt.subplots(stack,2,sharex=True,figsize=[11,6])
        fig.subplots_adjust(hspace=0)
        ax[1,0].axvline((z[1]/1e10+1.),
                     c=line_color, alpha=alpha, ls='--', zorder=3,label='Sonic Point')
        ax[1,0].axvline(R_H,ls=':',c=line_color,label='Hill Radius')
    #Density
    ax[0,0].semilogy(radius, rho,c=line_color)
    ax[0,0].set_ylabel(r'Density (g/cm$^3$)')
    #Velocity (+Hill Radius and sonic point for legend purposes)
    ax[1,0].legend()
    ax[1,0].plot(radius, v/(10*100*1000),c=line_color)
    ax[1,0].set_ylabel(r'Velocity (10 km/s)')
    #Temperature
    r_crit = 1.+z[0]/1e10
    indx = len(radius[radius < r_crit])-1
    Mdot = (4*np.pi*r[indx]**2 *rho[indx]*v[indx])
    mdot = (Mdot/const.Msun)*3.154e+7 #in Msun/year
    print('*******Mdot = %.2e*********' %mdot)
    
    ax[0,1].semilogy(radius, T,c=line_color,label=r'$\dot{M}$ = %.1e M$_{\odot}$/yr' %(mdot))
    ax[0,1].set_ylabel(r'Temperature (K)')

    #Ionization fraction
    title = ''
    ax[1,1].set_prop_cycle(custom_cycler)
#     for j,spname in enumerate(species_list):
#         spname = spname.replace(' ','')
#         ax[1,1].semilogy(soln.soln_norm['r'],1-soln.soln['Ys_'+spname],label=spname,c=line_color)
#         title += spname+': %.2f, '%(soln.HX[j])
    ax[1,1].semilogy(radius,1-Ys_HI,label='HI',c=line_color)
    ax[1,1].semilogy(radius,1-Ys_HeI,label='HeI',c=line_color)
    ax[1,1].semilogy(radius,1-Ys_CI,label='CI',c=line_color)
#     ax[1,1].semilogy(radius,1-Ys_NI,label='NI',c=line_color)
#     ax[1,1].semilogy(radius,1-Ys_OI,label='OI',c=line_color)
#     ax[1,1].semilogy(radius,1-Ys_NeI,label='NeI',c=line_color)
    
    ax[1,1].set_ylabel(r'Ionization Fraction')
    ax[1,1].set_ylim((1e-3,1))
    ax[1,1].legend()
    
    #Plotting shared lines
    for k in range(2):
        ax[1,k].set_xlabel(r'Radius (R$_p$)')
        for m in range(stack):
            ax[m,k].axvline(z[1]/1e10+1.,
                             c=line_color, alpha=alpha, ls='--', zorder=3)
            ax[m,k].axvline(R_H,ls=':',c=line_color)
    plt.gca().set_xlim(left=Rmin)
    if alone==True:
        return ax,title
```
