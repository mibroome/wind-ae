```python
from wrapper.wrapper_utils.inputs import input_handler
from wrapper.wrapper_utils.windsoln import wind_solution
from wrapper import relax_wrapper
from subprocess import Popen, PIPE, STDOUT
import sys

class ramp_to_new_base:
    """
    Class containing relaxation result, object typically referred to as the
    planet. Also contains functions for analyzing relaxation result
    """
    def __init__(self, start_file='saves/windsoln.csv', expedite=False,
                 add_uservars=False):
        self.error = 0
        self.atomictable = pd.read_csv('McAstro/atoms/atomic_table.txt',comment='#')
        self.inputs = input_handler()
        with open(start_file, 'r') as f:
            i_spec = 0
            for line in f:
                line = re.findall(r'[^,\s]+', line)
                if line[0] == '#nspecies:':
                    self.nspecies = int(line[1:][0])
                elif line[0] == '#vars:':
                    self.varz = line[1:]
                elif line[0] == '#scales:':
                    line[1:] = [float(x) for x in line[1:]]
                    self.scales = line[1:]
                    self.scales_dict = {v:s for v, s in
                                        zip(self.varz, self.scales)}
                elif line[0] == "#plnt_prms:":
                    plnt_prms = [float(x) for x in line[1:]]
                    # planet parameters
                    self.Mp = plnt_prms[0]
                    self.Rp = plnt_prms[1]
                    self.Mstar = plnt_prms[2]
                    self.semimajor = plnt_prms[3]
                    self.Fuv = plnt_prms[4]
                    self.Lstar = plnt_prms[5]
                elif line[0] == "#phys_prms:":
                    phys_prms = [x for x in line[1:]]
                    # physics paramters
                    mass_fractions = np.zeros(self.nspecies)
                    atom_masses = np.zeros(self.nspecies)
                    species_names = []
                    for s in range(self.nspecies): #CHANGED
                        mass_fractions[s] = float(phys_prms[s])
                        species_names.append(phys_prms[self.nspecies+s]) 
                        atom_masses[s] = float(phys_prms[2*self.nspecies + s])
                    self.HX = mass_fractions
                    self.species_list = species_names
                    self.atomic_masses = atom_masses
                    self.molec_adjust = float(phys_prms[-1])
                elif line[0] == "#bcs:":
                    bcs = [float(x) for x in line[1:]]
                    # boundary conditions
                    self.Rmin = bcs[0]
                    self.Rmax = bcs[1]
                    self.rho_rmin = bcs[2]
                    self.T_rmin = bcs[3]
                    Ys_array = np.zeros(self.nspecies)
                    Ncol_array = np.zeros(self.nspecies)
                    for s in range(self.nspecies):
                        Ys_array[s] = float(bcs[4+s])
                        Ncol_array[s] = float(bcs[4+self.nspecies + s])
                    self.Ys_rmin = Ys_array
                    self.Ncol_sp = Ncol_array
                elif line[0] == "#tech:":
                    tech = [float(x) for x in line[1:]]
                    # tech parameters
                    self.breezeparam = tech[0]
                    self.rapidity = tech[1]
                    self.erfn = tech[2]
                    self.mach_limit = tech[3]
                elif line[0] == "#flags:":
                    flags = line[1:]
                    # flags
                    self.lyacool = int(flags[0])
                    self.tidalforce = int(flags[1])
                    self.bolo_heat_cool = float(flags[2])
                    self.integrate_outward = int(flags[3])
                elif line[0] == "#add_prms:":
                    add_prms = line[1:]
                    self.n_add_prms = int(add_prms[0])
                    self.add_param_name = [0]
                    self.add_param_val  = [0]
                    if self.n_add_prms>0:
                        add_param_name_array = []
                        add_param_val_array  = np.zeros(self.n_add_prms)
                        self.add_param_dict = {}
                        for i in range(self.n_add_prms):
                            add_param_name_array = np.append(add_param_name_array,add_prms[2*i])
                            add_param_val_array[i]  = float(add_prms[2*i+1])
                            self.add_param_dict[add_prms[2*i]] = float(add_prms[2*i+1])
                        self.add_param_name = add_param_name_array
                        self.add_param_val  = add_param_val_array
                elif line[0] == "#spec_prms:": #moved up to allow access to nspecies, may cause read problems
                    spec = [x for x in line[1:]]
                    self.npts = int(spec[0])
#                     self.nspecies = int(spec[1])
                    self.spec_date = spec[2]
                    self.spec_kind = spec[3]
                    self.spec_window = np.asarray([float(spec[4]),
                                                   float(spec[5])])
                    self.spec_resolved = np.asarray([float(spec[6]),
                                                     float(spec[7])])
                    self.spec_normalized = np.asarray([float(spec[8]),
                                                       float(spec[9])])
                    if (self.spec_kind != 'full' and
                        self.spec_window[0] != self.spec_window[1]):
                        print('WARNING: Source is monochromatic, '
                              'but window bounds not equal.\n'
                              f'         window: [{self.spec_window[0]}, '
                              f'{self.spec_window[1]}].')
                    self.ion_pot = np.zeros(self.nspecies)
                    for s in range(len(self.ion_pot)):
                        self.ion_pot[s] = float(spec[10+s])
                    sft = s
                    self.E_wl = np.zeros(self.npts)
                    self.wPhi_wl = np.zeros(self.npts)
                    self.sigma_wl = np.zeros((self.npts, self.nspecies))
                    self.species_spaced = (self.species_list).copy()
                elif line[0] == "##":
                    spec_data = [float(x) for x in line[1:]]
                    self.E_wl[i_spec] = spec_data[0]
                    self.wPhi_wl[i_spec] = spec_data[1]
                    for s in range(self.nspecies):
                        self.sigma_wl[i_spec][s] = spec_data[2+s]
                    i_spec += 1
                elif line[0][0] == "#": #Skipping other comments
                    continue
                else: # Assuming all comment strings are in header
                    if i_spec != self.npts:
                        print("ERROR: Failed reading simulations' spectrum.\n"
                              f'       Only read in {i_spec} points, '
                              f'but should have read in {self.npts} points.')
                    else:
                        word = ''
                        for s in range(self.nspecies):
                            word += f'sigma_{self.species_list[s]},'
                        self.sim_spectrum = pd.DataFrame(
                            np.column_stack((self.E_wl, self.wPhi_wl,
                                             self.sigma_wl)),
                            columns=['E', 'wPhi', *word.split(',')[:-1]])
                    break
        # Read in solution data as pandas data frame
        self.soln_norm = pd.read_csv(start_file, comment='#', header=None,
                                     names=self.varz)
        self.soln = self.soln_norm*self.scales
        # Make tuples of input files for ease of use
        self.planet_tuple = [self.Mp, self.Rp, self.Mstar, self.semimajor,
                             self.Fuv,self.Lstar]
        self.physics_tuple = [self.HX, self.species_list, self.molec_adjust, self.atomic_masses] 
        self.bcs_tuple = [self.Rmin, self.Rmax, self.rho_rmin, self.T_rmin,
                          self.Ys_rmin, self.Ncol_sp]
        self.add_tuple = [self.n_add_prms,self.add_param_name,self.add_param_val]
        self.flags_tuple = [self.lyacool, self.tidalforce,
                            self.bolo_heat_cool, self.integrate_outward]
        self.tech_tuple = [self.breezeparam, self.rapidity, self.erfn,
                           self.mach_limit]
        self.spectrum_tuple = [self.npts, self.nspecies, self.spec_date,
                               self.spec_kind, self.spec_window,
                               self.spec_resolved, self.spec_normalized,
                               self.ion_pot, self.E_wl, self.wPhi_wl,
                               self.sigma_wl, self.species_list]

    def rewrite_guess(self,regrid=False,degree=1):
        for i,sp in enumerate(self.species_list):
            self.species_list[i] = (sp).replace(' ','')
        g = open("inputs/guess.inp","w")
        #Header
        self.nspecies = len(self.Ys_rmin)
        g.write("#nspecies: %d\n" %self.nspecies)
        Ncol_string = ''
        Ys_string = ''
        string = ''
        floats = ''
        for n in range(self.nspecies):
            floats += '{:.17e},'
            string += '{:s},'
            Ncol_string += 'Ncol_'+self.species_list[n]+','
            Ys_string   += 'Ys_'+self.species_list[n]+','
        g.write("#vars: r,rho,v,T,"+Ys_string+Ncol_string+"q,z\n")
        g.write('#scales: '+(','.join('%.17e' %i for i in self.scales))+ '\n')
        g.write('#plnt_prms: '+(','.join('%.17e' %i for i in (self.planet_tuple)))+ '\n')
        g.write('#phys_prms: '+(floats+string+floats).format(*self.HX,*self.species_list,*self.atomic_masses)[:-1]
                +',{:.5f}\n'.format(self.molec_adjust))
        g.write('#bcs: '+(','.join('{:.17e}' for i in range(4+2*self.nspecies))).format(*self.bcs_tuple[:4],
                                                                                        *self.Ys_rmin,*self.Ncol_sp)+'\n')
        g.write('#tech: '+(','.join('%.17e' %i for i in self.tech_tuple))+ '\n')
        g.write('#flags: {:d},{:d},{:.5f},{:d}\n'.format(*self.flags_tuple))
        g.write('#add_prms: %d' %self.add_tuple[0])
        if self.n_add_prms > 0:
            for i in range(self.n_add_prms):
                g.write(',%s,%.17e'%(self.add_param_name[i],self.add_param_val[i]))
        g.write('\n')
        g.write('#spec_prms: {:d},{:d},'.format(self.npts,self.nspecies)+self.spec_date+','+self.spec_kind+
                ',{:.17e},{:.17e},{:.17e},{:.17e},{:.17e},{:.17e},'.format(*self.spec_window,*self.spec_resolved, 
                                                                           *self.spec_normalized))
        g.write(floats[:-1].format(*self.ion_pot)+'\n')   
        #Spectrum
        #just adds new species at this stage. In other functions, can expand or change range,norm,etc.
        g.write("# $hc/\lambda_i$, $w_i\Phi_{\lambda_i}/F_{uv}$,"+
                (','.join("$\sigma_{\lambda_i,{"+sp+"}}$" for sp in self.species_list))+"\n")
        for n in range(len(self.E_wl)):
            g.write(('## {:.17e},{:.17e},'+floats[:-1]+'\n').format(self.E_wl[n],self.wPhi_wl[n],*self.sigma_wl[n]))
        #Guess Solution
        if regrid == False:
            for k in range(len(self.soln_norm)):
                g.write(','.join('%.17e' %i for i in self.soln_norm.iloc[k,:]))
                g.write('\n')
            g.close()

        r_og = self.soln_norm['r']
        q = self.soln['q']
        og_length = len(q[q<=1])
        percent_scaleheight = 0.02 
        if regrid == True:
            try_again = True
            while try_again == True:
                #Creating a new grid that is 2% of the scale height everywhere except as it approaches the sonic point
#                 sim.load_uservars('saves/windsoln.csv')
                little_g = const.G*self.Mp/(self.soln['r']**2)
                H_sc = const.kB*self.soln['T']/(self.soln['mu']*little_g)
                f_hsc = interpolate.CubicSpline(r_og[q<=1],H_sc[q<=1]/self.Rp)
                new_r = self.Rmin
                r_regrid = [new_r]
                r_steps = [1e-5]
                while new_r<self.R_sp:
                    r_step = f_hsc(new_r)*percent_scaleheight
                    r_steps = np.append(r_steps,r_step)
                    closeness = (self.R_sp-new_r) #if r close to Rsp, switch to an erf that decreases stepsize
                    if closeness<1:
                        r_step*=math.erf(closeness)+1e-4
                    new_r += r_step
                    r_regrid = np.append(r_regrid,new_r)  
                new_length = len(r_regrid) #the number of points inside the relaxation code
                if new_length < 1900: #randomly chosen length
                    try_again = False
                    r_regrid = np.append(r_regrid,r_og[q>1])
                else:
                    percent_scaleheight += 0.01 #can increase stepsize or smaller closeness


            rho_spline = interpolate.InterpolatedUnivariateSpline(self.soln_norm['r'], self.soln_norm['rho'], ext=2, k=degree)
            v_spline = interpolate.InterpolatedUnivariateSpline(self.soln_norm['r'], self.soln_norm['v'], ext=2, k=degree)
            T_spline = interpolate.InterpolatedUnivariateSpline(self.soln_norm['r'], self.soln_norm['T'], ext=2, k=degree)
            z_spline = interpolate.InterpolatedUnivariateSpline(self.soln_norm['r'], self.soln_norm['z'], ext=2, k=degree)
            z = self.soln_norm['z'][0]
            new_q = (r_regrid - self.Rmin)/z
            new_q[new_length-1] = 1

            def Ys_spline(r,degree=1):
                Ys_normed_fit = np.zeros((len(r),self.nspecies))
                for j in range(self.nspecies):
                    Ys_name = 'Ys_'+self.species_list[j]
                    s = interpolate.InterpolatedUnivariateSpline(
                        self.soln_norm['r'], self.soln_norm[Ys_name], ext=2, k=degree)
                    Ys_normed_fit[:,j] = s(r)
                return Ys_normed_fit

            def Ncol_spline(r,degree=1):
                Ncol_normed_fit = np.zeros((len(r),self.nspecies))
                for j in range(self.nspecies):
                    Ncol_name = 'Ncol_'+self.species_list[j]
                    s = interpolate.InterpolatedUnivariateSpline(
                        self.soln_norm['r'], self.soln_norm[Ncol_name], ext=2, k=degree)
                    Ncol_normed_fit[:,j] = s(r)
                return Ncol_normed_fit

            df = pd.DataFrame(np.column_stack((
                #strictly enforce r[0] == Rmin
                np.array(r_regrid),
                rho_spline(r_regrid),
                v_spline(r_regrid),
                T_spline(r_regrid),
                Ys_spline(r_regrid),
                Ncol_spline(r_regrid),
                new_q,
                z_spline(r_regrid)
            )), columns=self.soln_norm.columns[:(6+2*self.nspecies)])

            df.to_csv(g, header=False, index=False, float_format='%.17e')
            g.close()

            if og_length != new_length:
                print("Number of points in relaxation region has changed from %d to %d. Remaking C code..." 
                      %(og_length, new_length))
                h = (open('src/defs.h','r')).readlines()
                f = open('src/defs.h','w')
                for idx,hline in enumerate(h):
                    splitline = hline.split()
                    if len(splitline) >= 2:
                        line_var = splitline[0]+' '+splitline[1]
                    if line_var == '#define M':
                        index = idx
                h[index] = '#define M %d            /* number of points */\n' %new_length
                f.writelines(h)
                f.close()    
                sub = Popen('make', stdout=PIPE, stderr=PIPE) 
                output, error_output = sub.communicate()

        return


    def add_species(self, species_name):
        try:
            new_species = McAtom.formatting_species_list([species_name]) #adds spaces and converts, e.g., fe6 to Fe IX
        except IndexError:
            print("ERROR: Species must include an ionization number either in Roman or Arabic numerals (e.g., H1, heIV, Fe 6)", 
                  file=sys.stderr) # Python 3.x
            sys.exit(1)
        spec = spectrum()
        for sps in self.species_list:
            spec.add_species(McAtom.formatting_species_list([sps]))
        spec.set_resolved(*self.spec_resolved/1e-7) #
        spec.set_normalized(*self.spec_normalized/1e-7) 
        spec.set_window(*self.spec_window/1e-7,kind='full')
        spec.generate(kind='full', savefile='inputs/spectrum.inp')
        new_spec = np.genfromtxt('inputs/spectrum.inp',skip_header=8,delimiter=',')
        self.E_wl    = new_spec[:,0]
        self.wPhi_wl = new_spec[:,1]
        self.npts = len(self.wPhi_wl)
        self.sigma_wl = new_spec[:,2:]
        
        species = McAtom.formatting_species_list([species_name])
        ma = McAtom.atomic_species(species[0])
        sigma = ma.cross_section(self.E_wl/const.eV)
        sigma_array = np.zeros((self.npts, self.nspecies+1))
        for wl in range(len(self.E_wl)):
            sigma_array[wl] = np.append(self.sigma_wl[wl],[sigma[wl]])
        self.sigma_wl = sigma_array
        species_name = species[0].replace(' ','')
        
        prior_nspecies = len(self.Ys_rmin)
        self.soln_norm.insert(4+prior_nspecies, 'Ys_'+species_name, self.soln_norm.iloc[:,3+prior_nspecies], 
                              allow_duplicates=True) #takes last Ys column and duplicates it for the new species
        self.soln_norm.insert(5+2*prior_nspecies, 'Ncol_'+species_name, self.soln_norm.iloc[:,4+2*prior_nspecies]*1e-10, 
                              allow_duplicates=True) #takes last Ncol column and x1e-10 it for new species
        #Adding other species details
        self.HX = np.append(self.HX,0.) #0 because mass fraction is ramped up in another function
        self.ion_pot = np.append(self.ion_pot,ma.verner_data['E_th']*const.eV)
        self.atomic_masses = np.append(self.atomic_masses,
                                       self.atomictable['Atomic Mass'][self.atomictable['Name']==species_name])
        McAtom.formatting_species_list(self.species_list)
        tmp_list = np.append(self.species_list,species)
        for i in range(len(tmp_list)):
            tmp_list[i] = tmp_list[i].replace(' ','')
        self.species_list = tmp_list
        self.Ys_rmin = np.append(self.Ys_rmin,1.) #Neutral fraction
        self.Ncol_sp = np.append(self.Ncol_sp,1e-10) #0 mass fraction in the new species for now
        self.scales.insert(4+prior_nspecies,1.)
        self.scales.insert(5+2*prior_nspecies,1e17)
        
        self.nspecies = len(self.Ys_rmin) #reassigning new nspecies
        
        self.physics_tuple = [self.HX, self.species_list, self.molec_adjust,self.atomic_masses] 
        self.bcs_tuple = [self.Rmin, self.Rmax, self.rho_rmin, self.T_rmin,
                          self.Ys_rmin, self.Ncol_sp]
        self.flags_tuple = [self.lyacool, self.tidalforce,
                            self.bolo_heat_cool,
                            self.integrate_outward]
        self.tech_tuple = [self.breezeparam, self.rapidity, self.erfn,
                           self.mach_limit]
        self.spectrum_tuple = [self.npts, self.nspecies, self.spec_date,
                               self.spec_kind, self.spec_window,
                               self.spec_resolved, self.spec_normalized,
                               self.ion_pot, self.E_wl, self.wPhi_wl,
                               self.sigma_wl, self.species_list]
        
        #Changing Nspecies, M (# pts in relaxation region), RHOSCALE (convergence condition) and remaking C code    
        f = open('src/defs.h', 'r') 
        for line in f.readlines():
            splitline = line.split()
            if len(splitline) >= 2:
                line_var = splitline[0]+' '+splitline[1]
            if line_var == '#define NSPECIES':
                nspecies_def = int(line.split()[2])
            elif line_var == '#define M':
                og_length = int(line.split()[2]) 
            elif line_var == '#define RHOSCALE':
                og_RHOSCALE = float(line.split()[2])
        f.close()
        
        #rewriting
        new_length = len(self.soln['q'][self.soln['q']<=1])
        new_RHOSCALE = 10**np.floor(np.log10(self.rho_rmin*0.01))
        nspecies_new = int(self.nspecies)
        
        f = open('src/defs.h','w')
        h = (open('src/defs-master.h','r')).readlines()
        for idx,hline in enumerate(h):
            splitline = hline.split()
            if len(splitline) >= 2:
                line_var = splitline[0]+' '+splitline[1]
            if line_var == '#define NSPECIES':
                index1 = idx
            elif line_var == '#define M':
                index2 = idx
            elif line_var == '#define RHOSCALE':
                index3 = idx
        h[index1] = '#define NSPECIES %d\n' %nspecies_new
        h[index2] = '#define M %d            /* number of points */\n' %new_length
        h[index3] = '#define RHOSCALE %.1f\n' %new_RHOSCALE
        f.writelines(h)
        f.close()
        
        remake = False
        if nspecies_def != nspecies_new:
            print("Nspecies has changed from %d to %d. Remaking C code..." %(nspecies_def, nspecies_new))
            remake = True
        #For regridded solutions, the number of points in relaxtion region may change
        if og_length != new_length:
            print("Number of points in relaxation region has changed from %d to %d. Remaking C code..." 
                  %(og_length, new_length))
            remake=True
        #For solutions with very high rho at the lower boundary, the rho convergence condition should be raised
        if new_RHOSCALE != og_RHOSCALE:
            print("RHOSCALE (convergence condition) has changed from %d to %d. Remaking C code..." 
                  %(og_RHOSCALE, new_RHOSCALE)) 
            remake=True
        if remake == True:
            sub = Popen('make', stdout=PIPE, stderr=PIPE) 
            output, error_output = sub.communicate()
        return
                        
    def rewrite_all_inputs(self,integrate_out=False):
        # write input parameters
        self.inputs.write_planet_params(*self.planet_tuple)
        self.inputs.write_physics_params(*self.physics_tuple)
        self.inputs.write_bcs(*self.bcs_tuple)
        self.inputs.write_tech(*self.tech_tuple)
        self.inputs.write_flags(*self.flags_tuple,integrate_out=integrate_out)
        self.inputs.write_spectrum(*self.spectrum_tuple)
        return
                    
    def ramp_mass_frac(self,goal_mass_fracs,Ncol_sp_tot,integrate_out=False):
        '''mass_frac = array_like; mass fraction in each species'''
        if len(goal_mass_fracs) != len(self.species_list):
            sys.exit("ERROR: Mass fraction and species list must be the same length.")
        if np.round(np.sum(goal_mass_fracs),5) != 1:
            print('WARNING: Total Mass Fraction must sum to 1. sum(ZX) = %.3f' %np.sum(goal_mass_fracs))
        for i,sp in enumerate(self.species_list):
            self.species_list[i] = (sp).replace(' ','')
        
        self.HX = np.array(goal_mass_fracs)
        self.Ncol_sp = Ncol_sp_tot*self.HX
        
        self.physics_tuple = [self.HX, self.species_list,self.molec_adjust, self.atomic_masses] 
        self.bcs_tuple = [self.Rmin, self.Rmax, self.rho_rmin, self.T_rmin,
                          self.Ys_rmin, self.Ncol_sp]
        
        self.rewrite_all_inputs(integrate_out=integrate_out)
        self.rewrite_guess() #does not change columns of guess, because just has to be close enough
#         out = self.run_wind()
#         if out == 1:
#             print("The guess is ramping to itself, so that shouldn't happen!")
#         elif out == 2:
#             print("Try a smaller jump.")
        return  

    def FourPlot(self,soln,line_color='k',line_style='-',alone=True,ax=0): 
        radius_norm = 1.0
        alpha=0.5
        radius =soln.soln_norm['r']
        minrad = float(radius[soln.soln['T'] == np.min(soln.soln['T'])])
        R_H =soln.semimajor*(soln.Mp/(3*soln.Mstar))**(1/3) /self.Rp

        custom_cycler = (cycler(linestyle=['-', '--', ':', '-.',(0, (1, 10))]))

        stack=2
        if alone==True:
            fig, ax = plt.subplots(stack,2,sharex=True,figsize=[11,6])
            fig.subplots_adjust(hspace=0)
            ax[1,0].axvline(radius_norm*(soln.soln_norm['z'][1]+1.),
                         c=line_color, alpha=alpha, ls='--', zorder=3,label='Sonic Point')
            ax[1,0].axvline(R_H,ls=':',c=line_color,label='Hill Radius')
        #Density
        ax[0,0].semilogy(soln.soln_norm['r'],soln.soln['rho'],c=line_color)
        ax[0,0].set_ylabel(r'Density (g/cm$^3$)')
        #Velocity (+Hill Radius and sonic point for legend purposes)
        ax[1,0].legend()
    #     ax[0,0].set_ylim((1e-18,1e-10))
        ax[1,0].plot(soln.soln_norm['r'],soln.soln['v']/(10*100*1000),c=line_color)
        ax[1,0].set_ylabel(r'Velocity (10 km/s)')
    #     ax[1,0].set_ylim((0,10))
        #Temperature
        mdot = soln.Mdot #in Msun/year
        print('*******Mdot = %.2e g/s*********' %mdot)
        ax[0,1].semilogy(soln.soln_norm['r'],soln.soln['T'],c=line_color,label=r'$\dot{M}$ = %.2e g/s' %(mdot))
        ax[0,1].set_ylabel(r'Temperature (K)')
        ax[0,1].legend()
    #     ax[2,0].set_ylim((5e2,5e4))
        #Ionization fraction
        title = ''
        ax[1,1].set_prop_cycle(custom_cycler)
        for j,spname in enumerate(soln.species_list):
            spname = spname.replace(' ','')
            ax[1,1].semilogy(soln.soln_norm['r'],1-soln.soln['Ys_'+spname],label=spname,c=line_color)
            title += spname+': %.2f, '%(soln.HX[j])
        ax[1,1].set_ylabel(r'Ionization Fraction')
        ax[1,1].set_ylim((1e-3,1))
        ax[1,1].legend()
        # ax[2,1].set_ylim((1e12,1ffe22))
        #Plotting shared lines
        for k in range(2):
            ax[1,k].set_xlabel(r'Radius (R$_p$)')
            for m in range(stack):
                ax[m,k].axvline(radius_norm*(soln.soln_norm['z'][1]+1.),
                                 c=line_color, alpha=alpha, ls='--', zorder=3)
                ax[m,k].axvline(R_H,ls=':',c=line_color)
        plt.gca().set_xlim(left=soln.Rmin)
        if alone==True:
            return ax,title
    
    def run_wind(self, retry=False, expedite=False, verbose=False):
        # Run wind relaxation code
        sub = Popen('./bin/relaxed_ae', stdout=PIPE, stderr=PIPE) 
        output, error_output = sub.communicate()
        if error_output:
            if verbose:
                print(error_output)
            return 1
        else:
            # If successful then update our guess to new solution
            sub = Popen(["cp", 'saves/windsoln.csv', 'inputs/guess.inp'],
                        stdout=PIPE, stderr=PIPE)
            output, error_output = sub.communicate()
            if error_output:
                print(error_output)
                return 2
            self.windsoln = wind_solution(expedite=expedite)
            if self.windsoln.error:
                # If we ran into analysis errors and not runtime errors, then
                # we probaby need to converge the bcs.
                if not retry:
                    # Check retry to avoid infinite recursion
#                     self.converge_all_bcs(expedite=expedite)
                    self.ramp_base_bcs(expedite=expedite)
                    return self.run_wind(retry=True)
                else:
                    print("Failed on retry to fix analysis errors")
                    return 3
            # saves/windsoln.csv and inputs/guess.inp are the same
            self.guess = self.windsoln
            return 0

```

```python
def metallicity(species_list,Z=1):
    lodders = pd.read_csv('McAstro/atoms/Lodders.dat',comment='#',delimiter=' ')
    amu = pd.read_csv('McAstro/atoms/atomic_masses.csv', comment='#')
    
    try:
        species_list = McAtom.formatting_species_list(species_list) #adds spaces and converts, e.g., fe6 to Fe IX
    except IndexError:
        print("ERROR: Species must include an ionization number either in Roman or Arabic numerals (e.g., H1, heIV, Fe 6)", 
              file=sys.stderr) # Python 3.x
        sys.exit(1)

    N_per_N_H = 10**(lodders['A']-12)
    mn = np.zeros_like(N_per_N_H)
    for i,el in enumerate(lodders['name']):
        mass = (amu['mass'][amu['Symbol'] == el]).iloc[0]
        if (el != 'H') and (el != 'He'):
            mn[i] = mass*Z*N_per_N_H[i]
        else:
            mn[i] = mass*N_per_N_H[i]
    solar_mass_frac = mn/np.sum(mn)
    el_list = [species.split()[0] for species in species_list]
#     return solar_mass_frac[lodders.name.isin(el_list)]
    Z_array = []
    for element in el_list:
        Z_array = np.append(Z_array,solar_mass_frac[lodders['name']==element])
    temp_delta = (1-np.sum(Z_array))
    Z_array[0] += temp_delta
    return Z_array
```

```python
species = ['H I','He I','O I','C I','N I','Ne I','Mg I']

def add_metals(start_file,desired_species,Z,Ncol_sp):
    McAtom.formatting_species_list()
    #if desired_species[0] != 'H I':
    mfs = [metallicity('H I',Z)]
    ramp = ramp_to_new_base(start_file)
    for i,sp in enumerate(desired_species):
        ramp.add_species(sp)
        ramp.rewrite_guess(regrid=False)
        ramp.rewrite_all_inputs()
        
        mfs = np.append(mfs,metallicity(sp,Z)[0])
        temp_delta = (1-np.sum(mfs))
        mfs[0] += temp_delta
        ramp.ramp_mass_frac(mfs,Ncol_sp_tot=Ncol_sp)
        fail = 1
        goal_mf = mfs[-1]
        while ramp.run_wind(verbose=True) == 1:
            current_mf = goal_mf/(2**fail)
            print(f'\rRamping up mass fraction of new species: {current_mf:.2e}, goal: {goal_mf:.2e}',
                  end='                                                                              ')
            mfs[-1] = current_mf
            ramp.ramp_mass_frac(mfs,Ncol_sp_tot=Ncol_sp)
            if fail>10:
                break
            fail+=1
            
        print(f'\r Sucess! {sp:s} has been sucessfully ramped to {Z:d}xZ',
             end='                                                        ')
        ramp = ramp_to_new_base(start_file='saves/windsoln.csv')

# add_metals('saves/HD209458b_10-8000eV_H_mbar.csv',extra_species,1,0.11)
```
