```python
# E=np.linspace(20,120,100)
def Kshell_sigma(species,E_th,Energy,units='cm^2'):
    '''Gives Kshell scattering cross section in cm^2 or Mb as a function of energy in eV.
    
    Inputs: Z - int; atomic number 
            N - int; Number of electrons
            E_th - float; **K-shell** ionization potential in eV
            Energy - float or array; energy in eV of incident photon
            units - 'cm^2' or 'Mb'
    Returns: sigma - float or array; K shell scattering cross section
    '''
    ma = McAtom.atomic_species(species)
    Z = ma.Z; N = ma.Ne
    
    a1,a2,a3,a4,a5 = [11.132,5.612,6.9668,8.9690,0.35327]
    b1,b2,b3,b4    = [1.6798e-2,3.6703e-4,9.7881e-4,8.0332e-3]

    lowZs = np.array([[2, 2, 112.88, 3.9492],
    [3, 3, 144.47, 9.3820],
    [3, 2, 201.22, 3.3064],
    [4, 4, 184.11, 13.931],
    [4, 3, 175.67, 15.889],
    [4, 2, 200.76, 10.343]])

    if (Z in lowZs[:,0]) & (N in lowZs[:,1]):
        E_0,sig_0 = lowZs[(lowZs[:,0]==Z) & (lowZs[:,1]==N)][0][2:]
    else:
        E_0 = a1 + a2*Z + a3*Z**2 + (a4+a5*Z)*N #in eV
        sig_0 = 1/(b1 + b2*Z + b3*Z**2 + b4*N) #in Mb (x1e-18 to get to cm^2)

    y   = E / E_0 
    F   = y**(-1.5) * (1+y**(0.5))**(-4)
    print('F',F)
    if units == 'cm^2':
        scale = 1e-18
    elif units == 'Mb':
        scale = 1
    sigma = sig_0*F*scale
    sigma[E<E_th] = 0
    return sigma
```

```python
species = ['C I']
species = McAtom.formatting_species_list(species)[0]
T = np.arange(1,100e3)
temp = 1e4

ma = McAtom.atomic_species(species)
Verner = pd.read_csv('McAstro/atoms/Verner.csv', comment='#')
#IONIZATION POTENTIAL NEEDS TO BE K SHELL ION POT
# I = (Verner['E_th'][(Verner['Ne']==ma.Ne) & (Verner['Z']==ma.Z)]).item() #ionization potential of C in ergs
Hartree = 27.211296 #eV
I = 11.34*Hartree
print(I)
r_spec = 0.174 #radius of maximum density of orbital (Desclaux 1973) in 
N = 2          #number of electrons in orbital, always two for K Shell
element = species.split()[0]
if element in ['C','N','O']:
    al,bl,cl,dl = [1.75,1,1,1]
else:
    al,bl,cl,dl = [1.06,0.23,1,1.1]    

r = r_spec*5.29e-9#1e-7
J = const.me*const.c**2/(I*const.eV)
print('J',J)
v = np.sqrt(3*const.kB*temp/const.me) #electron thermal v, chose as mean of 3D
# print(const.me)
# T = const.me * v**2 / 2 #KE of electron in ergs
t = T/I
g_nl = 70/(I) #weighting factor for (n,l) orbital. g_1s = 70/I  # I bet in eV

f = dl*((t-1)/(t+1))**al * (bl + cl*(1-0.5/t)*np.log(2.7+(t-1)**0.5))
half = ( ((1+t)*(t+2*J)*(1+J)**2) / (J**2*(1+2*J) + t*(t+2*J)*(1+J)**2) )**1.5
F_rel = ((1+2*J)/(t + 2*J)) * ((t+J)/(1+J))**2 * (1 + 2*t**0.25 / J**2) * half
sigma_K = np.pi * r**2 * g_nl * (1/t) * f * N * F_rel
# print(sigma_K[50])

power = np.round(np.log10(np.nanmax(sigma_K)))-1
power = -20
plt.plot(T/1000,sigma_K/10**power)
plt.xlabel('Incident Kinetic Energy [keV]')
plt.ylabel(r'$\sigma$ [10$^{%.0f}$ cm$^2$]' %power)
plt.xscale('log')
# plt.xlim((0,100))
plt.show()
```
