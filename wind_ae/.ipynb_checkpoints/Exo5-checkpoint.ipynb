```python
def diffs(sim1,sim2):
    goal = sim2.windsoln.Mdot
    old = sim1.windsoln.Mdot
    percent = 100*abs(old-goal)/goal
    rH_idx1 = np.searchsorted(sim1.windsoln.soln_norm['r'],
                             sim1.windsoln.rHill)
    rH_idx2 = np.searchsorted(sim2.windsoln.soln_norm['r'],
                             sim2.windsoln.rHill)
    deltaT = sim1.windsoln.soln['T'][rH_idx1]-sim2.windsoln.soln['T'][rH_idx2]
    delta_f_H = sim1.windsoln.soln['Ys_HI'][rH_idx1]-sim2.windsoln.soln['Ys_HI'][rH_idx2]
#     delta_f_He = sim1.windsoln.soln['Ys_HeI'][rH_idx1]-sim2.windsoln.soln['Ys_HeI'][rH_idx2]
    print("Mdot: %.0f, %.2e old, %.2e new"%(percent,old,goal))
    print("T: %.0f"%deltaT)
    print("f: %.2f"%delta_f_H) 

def plot_tau(sim,color,eV,ls='-'):
    r = sim.windsoln.soln_norm['r']
    idx = len(sim.windsoln.E_wl) - np.searchsorted(np.flip(sim.windsoln.E_wl/const.eV),eV)-1
    tau = sim.windsoln.soln['Ncol_HI']*sim.windsoln.sigma_wl[idx][0]
    tau_idx = len(tau) - np.searchsorted(np.flip(tau),1)
    plt.axvline(r[tau_idx],c=color,ls=ls)
    return
```
