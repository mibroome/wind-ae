```python
sim.load_planet('saves/for_paper/new/starts/54.99Me_1.85Re.csv',expedite=False)
# sim.load_uservars('saves/for_paper/new/HD189_13.6-2000eV_H_polished_hd189spec.csv')

heat = sim.windsoln.soln['heat_ion']
cool = sim.windsoln.soln['cool_PdV']
drop_index = np.where(heat>-cool)[0][0]

def T_iterator(step,direction=1):
    last_mean = np.copy(abs(np.mean(np.diff(sim.windsoln.soln['T'])[:2])))
    sim.save_planet('saves/last.csv',overwrite=True)
    sim.ramp_T_rmin(sim.windsoln.T_rmin+step*direction)
    current_mean = abs(np.mean(np.diff(sim.windsoln.soln['T'])[:2]))
    T_scale = sim.windsoln.scales_dict['T']

    while current_mean < last_mean:
        last_mean = current_mean
        sim.save_planet('saves/last.csv',overwrite=True)
        sim.ramp_T_rmin(sim.windsoln.T_rmin+step*direction)
        current_mean = abs(np.mean(np.diff(sim.windsoln.soln['T'])[:2]))
        print(f'\r Last: {last_mean*T_scale:.0f}, Current: {current_mean*T_scale:.0f}',
              end='                                                                  ')
    sim.load_planet('saves/last.csv',print_atmo_composition=False)
    return

if drop_index < 10:
    delta_T = np.diff(sim.windsoln.soln['T'])
    oscill_count = 0
    for i in range(1,len(delta_T)): #if slope of temperature oscillates about 0
        if (delta_T[i-1] < -1) & (delta_T[i] > 1):
            oscill_count += 1
        if (delta_T[i-1] > 1) & (delta_T[i] < -1):
            oscill_count += 1
    if oscill_count > 5: #if it oscillates signficantly, try raising T in steps of 1000        
        T_iterator(0.1,1) #steps of 1000 up
            
        sim.ramp_T_rmin(sim.windsoln.T_rmin+0.01)
        up = abs(np.mean(np.diff(sim.windsoln.soln['T'])[:2]))
        sim.ramp_T_rmin(sim.windsoln.T_rmin-0.02)
        down = abs(np.mean(np.diff(sim.windsoln.soln['T'])[:2]))
        
        if up < down:
            T_iterator(0.01,1)
        else:
            T_iterator(0.01,-1)
        
        print('Final T: ',sim.windsoln.T_rmin*T_scale)  
    
# abs(np.mean(np.diff(sim_og.windsoln.soln['T'])[:2]))
```
