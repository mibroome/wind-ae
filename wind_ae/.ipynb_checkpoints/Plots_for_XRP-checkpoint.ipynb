```python
# Common python packages
import os
import re
from scipy import integrate
from scipy.interpolate import CubicSpline
import matplotlib
import matplotlib.pyplot as plt
import astropy.constants as c
import astropy.units as u
import warnings
from matplotlib.lines import Line2D
from cycler import cycler


cc = plt.rcParams['axes.prop_cycle'].by_key()['color']

params = {
'xtick.direction': 'in',
'ytick.direction': 'in',
}
plt.rcParams.update(params)
matplotlib.rc('font', **{'family': 'serif', 'serif': ['Computer Modern'],'size':15})
plt.rc('text', usetex=True)
warnings.filterwarnings("ignore")
%config InlineBackend.figure_format='retina'

# Check that pwd is relaxation directory and code is compiled
if os.path.exists(os.getcwd()+'/wrapper_tutorial.ipynb'):
    os.chdir('../..') # May need to be set by hand if nonconventional setup
if not os.path.exists(os.getcwd()+'/bin/relaxed_ae'):
    print('ERROR: relaxation code not found!\n'
          '       Either not compiled, or not in relaxation code directory')
    assert(False)

# Wrapper modules
from wrapper.relax_wrapper import wind_simulation as wind_sim
from wrapper.wrapper_utils import constants as const
from wrapper.wrapper_utils.system import system
from wrapper.wrapper_utils.physics import physics
from wrapper.wrapper_utils.spectrum import spectrum
from wrapper.wrapper_utils.plots import *
from wrapper.comparison_plots import ComparisonPlots
import McAstro.atoms.atomic_species as McAtom
# from massradius import get_Rp

def ergs_to_wlcm(energy):
    wl_nm = 1240 / (6.242e+11*energy)
    wl_cm = wl_nm*1e-7
    return wl_cm

def ergs_to_wlnm(energy):
    wl_nm = 1240 / (6.242e+11*energy)
    return wl_nm

path = '../../../mnt/c/Users/M/Documents/2_UCSC/Research/Figures/Good_xray/'
alpha=0.5

def tau_finder(E_input,sim_object):
    E_ev = sim_object.windsoln.E_wl*6.24e11
    delta_E = E_ev-E_input
    try:
        E_idx = np.where(np.abs(delta_E)==np.min(np.abs(delta_E)))[0][0]
        nspecies = len(sim_object.windsoln.sigma_wl[0])
#         taus = np.genfromtxt('taus.txt')[:,E_idx]
        taus = np.zeros_like(sim_object.windsoln.soln['r'])
        for i,col_name in enumerate(sim_object.windsoln.soln.columns[4+nspecies:4+2*nspecies]):
#             taus = simwindsoln.soln['Ncol_HI']*simwindsoln.sigma_wl[E_idx,0]+ simwindsoln.soln['Ncol_HeI']*simwindsoln.sigma_wl[E_idx,1]
            taus += sim_object.windsoln.soln[col_name]*sim_object.windsoln.sigma_wl[E_idx,i]
        tau1 = np.flip(np.array(r_norm[taus>=1]))[0]
        return tau1,E_ev[E_idx]
    except IndexError:
        return np.nan,E_ev[np.where(delta_E==np.min(np.abs(delta_E)))[0]]
```

```python
def plot_tau1_surfs(sim_object,label_height,fontsize=10):
    E_ev = sim_object.windsoln.sim_spectrum['E']/1.6e-12
    r_tau = []
    for E in E_ev:
        rt = tau_finder(E,sim_object)
        if np.isnan(rt[0])==False:
            r_tau.append(rt)
    color = plt.cm.magma(np.linspace(0, 1, len(r_tau)))
    [plt.axvline(r_tau[2*i][0],ls=':',c=color[2*i]) for i in range(int(len(r_tau)/2))]
    [plt.text(r_tau[4*i][0]-0.01,label_height+0.02e-6,'%.1f'%r_tau[4*i][1],fontsize=fontsize,backgroundcolor='w') for i in range(int(len(r_tau)/4))]
    [plt.text(r_tau[2+4*i][0]-0.01,label_height+0.05e-6,'%.1f'%r_tau[2+4*i][1],fontsize=fontsize,backgroundcolor='w') for i in range(int(len(r_tau)/4))]
```

```python
def SixPlot(planet_path='saves/windsoln.csv',line_color='k',line_style='-',alone=True,ax=0):
    sim = wind_sim()
    sim.load_planet(planet_path)
    sim.load_uservars()
    soln = sim.windsoln
    
    radius_norm = 1.0
    alpha=0.5
    radius = soln.soln_norm['r']
    minrad = float(radius[soln.soln['T'] == np.min(soln.soln['T'])])
    R_H = soln.semimajor*(soln.Mp/(3*soln.Mstar))**(1/3) / soln.Rp

    stack=3
    if alone==True:
        fig, ax = plt.subplots(stack,2,sharex=True,figsize=[12,10])
        fig.subplots_adjust(hspace=0)
        ax[1,0].axvline(radius_norm*(soln.soln_norm['z'][1]+1.),
                     c=line_color, alpha=alpha, ls='--', zorder=3,label='Sonic Point')
        ax[1,0].axvline(R_H,ls=':',c=line_color,label='Hill Radius')
    #Density
    ax[0,0].semilogy(soln.soln_norm['r'], soln.soln['rho'],c=line_color)
    ax[0,0].set_ylabel(r'$\rho$ (g/cm$^3$)')
    #Velocity (+Hill Radius and sonic point for legend purposes)
    ax[1,0].legend()
#     ax[0,0].set_ylim((1e-18,1e-10))
    ax[1,0].plot(soln.soln_norm['r'], soln.soln['v']/(10*100*1000),c=line_color)
    ax[1,0].set_ylabel(r'$v$ (10 km/s)')
#     ax[1,0].set_ylim((0,10))
    #Temperature
    ax[2,0].semilogy(soln.soln_norm['r'], soln.soln['T'],c=line_color)
    ax[2,0].set_ylabel(r'T (K)')
#     ax[2,0].set_ylim((5e2,5e4))
    #Number density
    ax[0,1].semilogy(soln.soln_norm['r'],soln.soln['n_H'],label='n',ls='-',c=line_color)
    ax[0,1].semilogy(soln.soln_norm['r'],soln.soln['n_HI'],label=r'n$_0$',ls='--',c=line_color)
    ax[0,1].semilogy(soln.soln_norm['r'],soln.soln['n_HII'],label=r'n$_+$',ls=':',c=line_color)
    ax[0,1].set_ylabel(r'n (cm$^{-3}$)')
    if alone == True:
        ax[0,1].legend()
    custom_cycler = (cycler(linestyle=['-', '--', ':', '-.',(0, (1, 10))]))
#         custom_cycler=plt.cycler("color", plt.cm.Greys_r(np.linspace(0,1,soln.nspecies+1)))
#     else:
#         custom_cycler=plt.cycler("color", plt.cm.Blues_r(np.linspace(0,1,soln.nspecies+1)))
    #Ionization fraction
    title = ''
    ax[1,1].set_prop_cycle(custom_cycler)
    for j,spname in enumerate(soln.species_list):
        ax[1,1].semilogy(soln.soln_norm['r'],1-soln.soln['Ys_'+spname],label=spname,c=line_color)
        title += spname+': %.2f, '%(soln.HX[j])
    ax[1,1].set_ylabel(r'f$_+$ = n$_+$/n')
    ax[1,1].set_ylim((1e-3,1))
    ax[1,1].legend()
    #Column Density
#     custom_cycler = (cycler(color=['c', 'm', 'y', 'k']) +
#                  cycler(lw=[1, 2, 3, 4]))
    ax[2,1].set_prop_cycle(custom_cycler)
    for spname in soln.species_list:
        ax[2,1].semilogy(soln.soln_norm['r'],soln.soln['Ncol_'+spname],label=spname,c=line_color)
    ax[2,1].set_ylabel(r'Ncol (cm$^{-2}$)')
    ax[2,1].legend()
    # ax[2,1].set_ylim((1e12,1ffe22))
    #Plotting shared lines
    for k in range(2):
        ax[2,k].set_xlabel(r'Radius (R$_p$)')
        for m in range(stack):
            ax[m,k].axvline(radius_norm*(soln.soln_norm['z'][1]+1.),
                             c=line_color, alpha=alpha, ls='--', zorder=3)
            ax[m,k].axvline(R_H,ls=':',c=line_color)
    plt.gca().set_xlim(left=soln.Rmin)
    if alone==True:
        return ax,title

def ComparativeSixPlot(planet1_tuple, planet2_tuple,title1='',title2='',savepath=''):
    ax,title=SixPlot(*planet1_tuple,alone=True)
    SixPlot(*planet2_tuple,alone=False,ax=ax)
    custom_lines = [Line2D([0], [0], color=planet1_tuple[1], lw=2),
                    Line2D([0], [0], color=planet2_tuple[1], lw=2)]
    if (title1 == '') & (title2 == ''):
        title1 = (((planet1_tuple[0]).split('/'))[-1].split('.')[0]).replace('_',' ')
        title2 = (((planet2_tuple[0]).split('/'))[-1].split('.')[0]).replace('_',' ')
    ax[0,0].legend(custom_lines, [title1,title2])
#     fig.tight_layout()
    if savepath != '':
        plt.savefig(savepath)
    plt.show()
    return
```
