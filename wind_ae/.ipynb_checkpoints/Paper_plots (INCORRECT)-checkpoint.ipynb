#### 2. Multifrequency energy + tau <a id='energy'></a>

```python
def tau_finder(E_input,sim_object):
    E_ev = sim_object.windsoln.E_wl*6.24e11
    delta_E = E_ev-E_input
    r_norm = sim_object.windsoln.soln_norm['r']
    try:
        E_idx = np.where(np.abs(delta_E)==np.min(np.abs(delta_E)))[0][0]
        nspecies = len(sim_object.windsoln.sigma_wl[0])
#         taus = np.genfromtxt('taus.txt')[:,E_idx]
        taus = np.zeros_like(sim_object.windsoln.soln['r'])
        for i,col_name in enumerate(sim_object.windsoln.soln.columns[4+nspecies:4+2*nspecies]):
#             taus = simwindsoln.soln['Ncol_HI']*simwindsoln.sigma_wl[E_idx,0]+ simwindsoln.soln['Ncol_HeI']*simwindsoln.sigma_wl[E_idx,1]
            taus += sim_object.windsoln.soln[col_name]*sim_object.windsoln.sigma_wl[E_idx,i]
        tau1 = np.flip(np.array(r_norm[taus>=1]))[0]
        return tau1,E_ev[E_idx]
    except IndexError:
        return np.nan,E_ev[np.where(delta_E==np.min(np.abs(delta_E)))[0]]
```

```python
sim2 = wind_sim()
sim3 = wind_sim()

sim2.load_planet('saves/HD209458b/HD209458b_mono20eV_H.csv',
                expedite=False)
sim3.load_planet('saves/HD209458b/HD209458b_H_13.6-2000eV.csv',
                expedite=False)

fig,ax=plt.subplots(figsize=(8,6))
energy_plot(sim3.windsoln,ax=ax)
energy_plot(sim2.windsoln, ax=ax,alpha=0.3,legend=False)
tau3 = tau_finder(20,sim3)[0]
tau2 = tau_finder(20,sim2)[0]
plt.axvline(tau2,label='20 eV Mono',c='tab:gray',ls='--')
plt.axvline(tau3,label='20 eV Multi',c='k')
plt.text(1.5,6e-7,r'Multi.$\tau$(20eV)=1',horizontalalignment='center',
     verticalalignment='center')
plt.text(1.15,6e-7,r'Mono. $\tau$(20eV)=1',horizontalalignment='center',
     verticalalignment='center')


plt.xlim(1,2.1)
# plt.legend(loc='lower left',bbox_to_anchor=(1.1,0),fontsize=12)
plt.savefig('../Figures/temp_energy_plot.pdf',dpi=500,bbox_to_anchor='tight',
            facecolor='white')
plt.show()
```

Acceptable: 0.05 - 1

```python
sim = wind_sim()
sim.load_planet('saves/WASP12b/WASP12b_H_He.csv',
                expedite=False)
sim.windsoln.semimajor/const.au
```

```python
nsteps = 2500
semimajor_array = np.linspace(0.0232,1,nsteps)
# print(semimajor_array)
energy_limit = (sim.windsoln.Ftot*np.pi*(sim.windsoln.Rp*1.2)**3 / (const.G*sim.windsoln.Mp))

for i,a in enumerate(semimajor_array[1:]):
    sim.check_isotherm()

    goal_F = 32.63241472/a**2
    print(f'\rSemimajor: {a:.4f}. Flux: {goal_F:.1f}.')
    if sim.ramp_var("Ftot",goal_F,converge_bcs=False,make_plot=False) == 0:
        print("Finished F!")
        last_planet = np.copy(sim.windsoln.planet_tuple)
        last_planet[3] = a*const.au
        sim.inputs.write_planet_params(*last_planet)
        if sim.run_wind() == 0:
            sim.ramp_base_bcs()
            sim.converge_Rmax()
            if i%(int(nsteps/20)) == 0:
                sim.save_planet(f'Semimajor/wasp12_{a:.2e}.csv')
    else:
        print("broken")
        break
```

### 7. Flux vs. Mp/Rp 

```python
Mp = [1.5 ,2.5 ,3.5, 4.5, 5.3, 5.5, 6.5 ,7.5, 8.5 ,9.5]
Rp = [1.7, 1.77, 1.84, 1.92, 1.99, 2.02, 2.06, 2.13, 2.21, 2.28, 2.35]
flux_list = np.linspace(35.46,3546,10)

sim = wind_sim()
Mdot_grid = np.zeros((len(flux_list),len(Mp)))
Kngrid = np.zeros((len(flux_list),len(Mp)))

for i,F in enumerate(flux_list):
    for j,M in enumerate(Mp):
        R = Rp[j]
        Fstr = str(int(np.round(F,0)))
        Mstr = str(np.round(M,1))
        Rstr = str(np.round(R,2))
        filename_str = 'toi776c_'+Mstr+'Me_'+Rstr+'Re_'+Fstr+'.csv'
        try:
            sim.load_uservars(f'saves/TOI776/toi776c/'+filename_str)
        except FileNotFoundError:
            Mdot_grid[i,j] = np.nan
            Kngrid[i,j] = np.nan
        Mdot_grid[i,j] = sim.windsoln.Mdot / 3
        Kngrid[i,j] = sim.windsoln.Kn_hb_crit
        
```

```python

plt.contourf(Mp,flux_list,Kngrid,levels=50)
plt.colorbar(label='Kn')

plt.show()
```

```python
# Mdot_grid = np.ma.array(Mdot_grid, mask=np.isnan(Mdot_grid))

fig,ax= plt.subplots(figsize=(6,4))
plt.contourf(Mp,flux_list,Mdot_grid/1e9,levels=50)
plt.plot()
plt.xlabel(r'M$_P$ ($ M_{Earth}$)')
plt.ylabel('Flux at %.2f au (ergs s$^{-1}$ cm$^{-2}$)'%(sim.windsoln.semimajor/const.au))
plt.colorbar(label='$\dot{M}$ ($10^9 g\ s^{-1}$)')
plt.contour(Mp,flux_list,Kngrid,levels=[1],colors='white')
ax.set_box_aspect(1)
plt.savefig('../Figures/temp_flux.pdf',dpi=500,
            facecolor='white')
plt.show()
```

```python
#attenuating the spectrum at 70eV to 70% of it's original flux
sim.windsoln('saves/HD209458b/HD209458b_H_13.6-2000eV.csv')
E = sim.windsoln.E_wl
flux_per_bin_star = E*sim.windsoln.Ftot*sim.windsoln.wPhi_wl
phi70 = np.copy(sim.windsoln.wPhi_wl)
F_EUV_star = sum(flux_per_bin_star[(E<100*const.eV)&(E>13.6*const.eV)])
print(F_EUV_star)
print(sim.windsoln.Ftot)
print(F_EUV_star/sim.windsoln.Ftot)
phi70[E>70*const.eV] = phi70[E>70*const.eV]*0.7
flux_per_bin_p = E*sim.windsoln.Ftot*phi70

F_EUV_p = sum(flux_per_bin_p[(E<100*const.eV)&(E>13.6*const.eV)])
F_tot_p = sum(flux_per_bin_p)
print(F_EUV_p)
print(F_tot_p)

plt.plot(E/const.eV,flux_per_bin_star,label='star')
plt.plot(E/const.eV,flux_per_bin_p,label='planet')
plt.axvline(165)
plt.xscale('log')
plt.show()

#writing attenuated spectrum to file        
sim.windsoln.spectrum_tuple[10] = phi70
sim.windsoln.planet_tuple[4]=F_tot_p
sim.inputs.write_spectrum(*sim.windsoln.spectrum_tuple)
sim.inputs.write_planet_params(*sim.windsoln.planet_tuple)
```
