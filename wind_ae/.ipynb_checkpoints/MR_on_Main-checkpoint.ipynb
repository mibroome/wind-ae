```python
Fstr = '1096'
for i in range(len(radii)):
    print(f'------\n M = {masses[i]/const.Mjupiter:.2f}, R = {radii[i]/const.Rjupiter:.2f}')
    sim.windsoln.planet_tuple[0] = masses[i]
    sim.windsoln.planet_tuple[1] = radii[i]
    result = sim.ramp_grav(system(*sim.windsoln.planet_tuple),
                           expedite=True,
                          integrate_out=False)
    if (result == 0) or (result==5):
        print("Success!")
        Mstr = str(np.round(sim.windsoln.Mp/const.Mjupiter,2))
        Rstr = str(np.round(sim.windsoln.Rp/const.Rjupiter,2))
        filename_str = Mstr+'Me_'+Rstr+f'Re_'+Fstr
        sim.save_planet('saves/_MRgrid/'+filename_str)
```

```python
sim = wind_sim()
sim.load_planet('saves/for_paper/super_earth_solar.csv')

#2e-4 to 4e-6
#times 0.85 to account for Sun being dimmer back then
Fmin = 4e-6*sim.windsoln.Lstar/(4*np.pi*(sim.windsoln.semimajor)**2)
Fmax = 0.85*2e-4*sim.windsoln.Lstar/(4*np.pi*(sim.windsoln.semimajor)**2)
print(Fmin, Fmax)

flux = np.linspace(Fmin,Fmax,20)
```

```python
for i in range(len(radii[:-4])):
    Mstr = str(np.round(masses[i+1]/const.Mjupiter,2))
    Rstr = str(np.round(radii[i+1]/const.Rjupiter,2))
    print("\n\nM="+Mstr)
    filename_str = Mstr+'Me_'+Rstr+f'Re_1096'
    sim.load_planet('saves/_MRgrid/'+filename_str+'.csv',print_atmo_composition=False)
    for F in flux:
        Fstr = str(int(F))
        print(f'-------\n{Fstr:s}')
        result = sim.ramp_var("Ftot",F,
                             converge_bcs=True,
                             expedite=True,
                             integrate_out=False)
        if (result == 0) or (result==5):
            print("Success!")
            Mstr = str(np.round(sim.windsoln.Mp/const.Mjupiter,2))
            Rstr = str(np.round(sim.windsoln.Rp/const.Rjupiter,2))
            filename_str = Mstr+'Me_'+Rstr+f'Re_'+Fstr
            print(filename_str)
            sim.save_planet('saves/_MRgrid/'+filename_str)
```

```python
mdots = np.zeros(len(radii),len(flux))
for i in range(len(radii[:-3])):
    Mstr = str(np.round(masses[i]/const.Mjupiter,2))
    Rstr = str(np.round(radii[i]/const.Rjupiter,2))
    for j,F in enumerate(flux):    
        Fstr = str(int(F))
        filename_str = Mstr+'Me_'+Rstr+f'Re_'+Fstr
        try:
            sim.load_uservars('saves/_MRgrid/'+filename_str+'.csv')
        except FileNotFoundError:
            mdots[i,j] = np.nan

        mdots[i,j] = sim.windsoln.Mdot
```
